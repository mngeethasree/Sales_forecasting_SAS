```python
df_us = df[df['Country'] == 'United States']
```

```python
# One order can have multiple products
df_us[df_us['Order ID']=='CA-2012-124891']
```

```python
# convert order date to date time variable type
df_us['Order Date'] = pd.to_datetime(df_us['Order Date'])

# Extract order month and order year for aggregation
df_us['Order Month'] = df_us['Order Date'].dt.month
df_us['Order Year'] = df_us['Order Date'].dt.year

#print unique years and months
print(df_us['Order Month'].value_counts())
print(df_us['Order Year'].value_counts())
```

```python
df_us.columns
```

```python
#extract necessary columns from data
df_us = df_us[['Row ID', 'Order ID', 'Order Date', 'Order Month', 'Order Year', 'Sales', 'Discount', 'Order Priority']]
```

```python
df_us['Order Priority'].value_counts()
```

```python
#Multiplying discount and sales to get discounted_sales for each product
df_us['Discounted_sales'] = df_us['Discount'] * df_us['Sales']
```

```python
# group by based on order_id and extract all the necessary variables
df_us_agg = df_us.groupby(['Order ID'],as_index=False).agg(
                                            sales = pd.NamedAgg(column='Sales', aggfunc='sum'),
                                            order_date = pd.NamedAgg(column='Order Date', aggfunc=lambda x: x.head(1)),
                                            Discount_avg = pd.NamedAgg(column='Discount', aggfunc='mean'),
                                            Discount_min = pd.NamedAgg(column='Discount', aggfunc='min'),
                                            Discounted_sales_sum = pd.NamedAgg(column='Discounted_sales', aggfunc='sum'),
                                            critical = pd.NamedAgg(column='Order Priority', aggfunc=lambda x: 1 if (x.iloc[0]=='Critical') else 0),
                                            high_priority = pd.NamedAgg(column='Order Priority', aggfunc=lambda x: 1 if (x.iloc[0]=='High') else 0),
                                            low_priority = pd.NamedAgg(column='Order Priority', aggfunc=lambda x: 1 if (x.iloc[0]=='Low') else 0),
                                            medium_priority = pd.NamedAgg(column='Order Priority', aggfunc=lambda x: 1 if (x.iloc[0]=='Medium') else 0),
                                            )
df_us_agg.head(5)
```

```python
df_us_agg.columns
```

```python
df_us_agg['Total_Discount_per_order'] = df_us_agg['Discounted_sales_sum']/df_us_agg['sales']
```

```python
df_us_agg.set_index('order_date', inplace=True)
monthly_agg = df_us_agg.resample('M').agg(sales=pd.NamedAgg(column='sales', aggfunc='sum'),
                                            Discount_avg = pd.NamedAgg(column='Discount_avg', aggfunc='mean'),
                                            Discount_min = pd.NamedAgg(column='Discount_min', aggfunc='min'),
                                            Total_Discount_per_order_avg = pd.NamedAgg(column='Total_Discount_per_order', aggfunc='mean'),
                                            critical_orders = pd.NamedAgg(column='critical', aggfunc='sum'),
                                            high_priority_orders = pd.NamedAgg(column='high_priority', aggfunc='sum'),
                                            low_priority_orders = pd.NamedAgg(column='low_priority', aggfunc='sum'),
                                            medium_priority_orders = pd.NamedAgg(column='medium_priority', aggfunc='sum'),
                                          )
monthly_agg.reset_index(inplace=True)
```

```python
monthly_agg.to_csv('/content/drive/MyDrive/OPIM_5671_Data_Mining/Forecasting_project/MONTHLY_SALES_discount.csv')
```
